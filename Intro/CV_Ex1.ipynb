```python
%matplotlib inline

import matplotlib.pyplot as plt
from imageio import imread # recommended py python
import numpy as np
```

## Image thresholding

(A) Load the 'plate.png' image.

(B) Plot its values histogram.

(C) Threshold the images at 4 diffrent values (th=10, 20, 160 and 170), and display the resulting binary images (use subplots).

```python
img = imread('plate.png')
plt.imshow(img,cmap="gray",interpolation="nearest")
plt.colorbar(fraction=0.025)
plt.show()
```

```python
plt.hist(img.flatten())
plt.show()
for i,v in enumerate([10,20,160,170]):
    plt.subplot(2,2,i+1).imshow(img > v,cmap="gray",interpolation="nearest")
plt.show()
```

### coins.jpg

### rice.jpg

(A) Load and display 'rice.jpg' image and its histogram.

(B) Threshold the image to produce a good binary segmentation (between the coins and its background).

(C) Calculate and compare the histograms of the top-half of the image and the bottom-half of the image.

```python
# your coimg_coins = imread('coins.jpg')
im_rice = imread('rice.jpg')
plt.imshow(im_rice,cmap="gray",interpolation="nearest")
plt.colorbar(fraction=0.025)
plt.show()

plt.hist(im_rice.flatten())
plt.show()
plt.imshow(im_rice > 120,cmap="gray",interpolation="nearest")
plt.show()
plt.subplot(1,2,1).hist(im_rice[:np.size(im_rice,0)//2,:].flatten()) # top
plt.subplot(1,2,2).hist(im_rice[np.size(im_rice,0)//2:,:].flatten()) # bottom
plt.show()


plt.subplot(1,2,1).imshow(im_rice[:np.size(im_rice,0)//2,:],cmap="gray",interpolation="nearest") # top
plt.subplot(1,2,2).imshow(im_rice[np.size(im_rice,0)//2:,:],cmap="gray",interpolation="nearest") # bottom
plt.show()
```
